**#Insight : This command counts the number of customers from each state. It helps identify regions with the highest customer concentration and market presence.**

```python
import matplotlib.pyplot as plt

customers['customer_state'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 States by Customers")
plt.show()
```

**#Insight : The chart shows which states have the highest customer concentration.**

```python
customers['customer_city'].value_counts().head(5)
```

**#Insight : Identifies cities with the highest number of customers.**

```python
customers['customer_city'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Cities by Customers")
plt.show()
```

**#Insight : Shows percentage distribution of customers across states.**

```python
customers['customer_state'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%')

plt.ylabel("")
plt.title("Top 5 States")
plt.show()
```

**#Insight : Visualizes customer share among top states.**

```python
import seaborn as sns

state_city = customers.groupby('customer_state')['customer_city'].count().reset_index()

sns.heatmap(state_city[['customer_city']].head(10), annot=True)

plt.title("Customer Density Heatmap")
plt.show()
```
